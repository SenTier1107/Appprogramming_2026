```python
from faker import Faker
import random

# Faker 객체 생성 (한국어 설정)
fake = Faker('ko_KR')

# 1. User 클래스 정의 (7가지 속성 + 딕셔너리 변환 추가)
class User:
    def __init__(self, name, age, address, phone_number, job, company, gender):
        self.name = name
        self.age = age
        self.address = address
        self.phone_number = phone_number
        self.job = job
        self.company = company
        self.gender = gender

    # 텍스트로 출력하기 위한 메서드
    def __repr__(self):
        return f"[{self.name} / {self.gender} / {self.age}세] 🏢 {self.company} | 💼 {self.job} | 📞 {self.phone_number} | 🏠 {self.address}"

    # 판다스 데이터프레임 변환을 위해 추가한 메서드
    def to_dict(self):
        return {
            "이름": self.name,
            "나이": self.age,
            "주소": self.address,
            "전화번호": self.phone_number,
            "직업": self.job,
            "회사": self.company,
            "성별": self.gender
        }

# 2. 1000명의 유저 객체를 만들어서 두 개의 리스트에 각각 담기
users = []       # 텍스트 출력용 (객체 저장)
users_data = []  # 판다스 표 출력용 (딕셔너리 저장)

for _ in range(1000):
    new_user = User(
        name=fake.name(),
        age=random.randint(20, 60),
        address=fake.address(),
        phone_number=fake.phone_number(),
        job=fake.job(),
        company=fake.company(),
        gender=random.choice(["남성", "여성"])
    )
    # 똑같은 데이터를 두 곳에 동시에 추가
    users.append(new_user)
    users_data.append(new_user.to_dict())

# 3. 데이터가 잘 만들어졌는지 맨 앞의 5명만 출력
print(f"총 생성된 유저 수: {len(users)}명\n")
for user in users[:5]:
    print(user)
```

```python
class User:
    # 1. 속성 초기화 (Initialize attributes)
    def __init__(self, name):
        self.name = name
        self.following = 0  # 내가 쫓아가는 횟수 (또는 사람 수)
        self.follower = 0   # 나를 쫓아오는 횟수 (또는 사람 수)

    # 2. 쫓아가는 행동 (Method)
    def following_mode(self, target):
        # 쫓아가는 사람의 following 수치 1 증가
        self.following += 1
        # 쫓기는 사람의 follower 수치 1 증가
        target.follower += 1

        print(f"🏃‍♂️ {self.name}이(가) {target.name}을(를) 쫓아갑니다!")
        print(f"   👉 상태 업데이트: [{self.name}] Following: {self.following} | [{target.name}] Follower: {target.follower}")

# ==========================================

# 1. Tom과 Jerry 객체(인스턴스) 생성
tom = User("Tom")
jerry = User("Jerry")

# 2. Tom이 Jerry를 10번 쫓아가는 반복문 실행
print("🚨 추격전을 시작합니다!\n")
for i in range(1, 11):
    print(f"[{i}번째 추격]")
    tom.following_mode(jerry)
    print("-" * 40)
```
